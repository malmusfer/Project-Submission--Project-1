# Exploring Weather Trends

```python
# Importing neccessary libraries for the project

import pandas as pd # for dealing with data!
import matplotlib.pyplot as plt # for visualizing the data!
import numpy as np# for calculating the moving average!
```

```python
#Importing the data

gt = pd.read_csv('global_temp.csv') # importing 'global tempreature data'
ct = pd.read_csv('Riyadh_temp.csv') # importing 'city tempreature data'
```

```python
#Moving averages are used to smooth out data and to make it easier to observe long term trends 
#and not get lost in daily fluctuations

gt['g_avg'] = gt['avg_temp'].rolling(10).mean()
ct['c_avg'] = ct['avg_temp'].rolling(10).mean()
```

```python
#Visualizng Data using matplotlib Library

plt.plot(gt['year'],gt['g_avg'],label='Global')
plt.plot(ct['year'],ct['c_avg'],label='Riyadh')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)") 
plt.title("Average Temperature for the Riyadh City")
plt.show()
```
